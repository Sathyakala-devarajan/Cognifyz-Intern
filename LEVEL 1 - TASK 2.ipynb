    1. Identify the city with the highest number of restaurants in the dataset.
    2. Calculate the average rating for restaurants in each city.
    3. Determine the city with the highest average rating.

```python
# Import library files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv('Dataset.csv')
```

```python
#Dispaly the table
data
```

# 1. Identify the city with the highest number of restaurants in the dataset.

```python
# Group by 'City' & count the number of restaurants in each city
city_count = data['City'].value_counts()
city_count
```

```python
# Get the city with the highest number of restaurants
city_by_High_restaurant = city_count.idxmax()

city_by_High_restaurant
```

```python
# Get maximum restaurants count by city
max_restaurants_count = city_count.max()
max_restaurants_count
```

```python
print(f"The city with the highest number of restaurants is '{city_by_High_restaurant}' with {max_restaurants_count} restaurants.")
```

# 2. Calculate the average rating for restaurants in each city.

```python
# calculate the average rating for each city
city_avg_ratings = data.groupby('City')['Aggregate rating'].mean().reset_index()

# Print the average ratings for each city
print(city_avg_ratings)
```

```python

# Create a bar graph to visualize the average ratings
plt.figure(figsize=(20, 6))
bars = plt.bar(city_avg_ratings['City'], city_avg_ratings['Aggregate rating'], color='orange')
plt.title('Average Ratings for Restaurants in Each City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)

# Display the average rating values in vertical orientation next to each bar
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                 textcoords="offset points", ha='center', va='bottom', rotation='vertical')

plt.tight_layout()

# Display the bar graph
plt.show()
```

# 3. Determine the city with the highest average rating.

```python
# Find the city with the highest average rating
city_with_highest_avg_rating = city_avg_ratings.loc[city_avg_ratings['Aggregate rating'].idxmax()]

print(city_with_highest_avg_rating)
```
