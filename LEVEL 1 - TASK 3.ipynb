# TASK -3: Price Range Distribution    1. Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
    2. Calculate the percentage of restaurants in each price range category.

```python
# Import library files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv('Dataset.csv')
data
```

```python
data.columns
```

# 1. Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
# Count the frequency of each price range
price_range_counts = data['Price range'].value_counts()

# Create a bar chart to visualize the distribution of price ranges
plt.figure(figsize=(8, 6))
bx = price_range_counts.plot(kind='bar', color='Green')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.3)
plt.grid(axis='x', alpha=0.3)
plt.xticks(rotation=0)

# Display the values near the bars
for i, v in enumerate(price_range_counts):
    bx.text(i, v + 5, str(v), ha='center', va='bottom')

plt.show()
```

# 2. Calculate the percentage of restaurants in each price range category.

```python

# Calculate the total number of restaurants
total_restaurants = len(data)

# Calculate the percentage of restaurants in each price range category
percentage_per_price_range = (price_range_counts / total_restaurants) * 100

# Display the percentages
print("Percentage of restaurants in each price range category:")
print(percentage_per_price_range)
```

```python
# Create a line chart to visualize the percentage
plt.figure(figsize=(10, 6))
percentage_per_price_range.plot(kind='line', marker='o', color='green')
plt.title('Percentage of Restaurants in Each Price Range Category')
plt.xlabel('Price Range Category')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.grid(True)

# Display the percentage values as labels on the data points
for x, y in zip(percentage_per_price_range.index, percentage_per_price_range):
    plt.text(x, y + 1, f'{y:.2f}%', ha='center', va='bottom')

plt.show()
```
