    1. Identify the restaurants with the highest and lowest number of votes.
    2. Analyze if there is a correlation between the number of votes and the rating of a restaurant.


```python
# Import library files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('Dataset.csv')
data
```

```python
data.columns
```

# 1. Identify the restaurants with the highest and lowest number of votes.

```python

# Find the restaurant with the highest number of votes
highest_votes = data[data['Votes'] == data['Votes'].max()]
highest_votes_name = highest_votes.iloc[0]['Restaurant Name']
highest_votes_count = highest_votes.iloc[0]['Votes']

# Find the restaurant with the lowest number of votes
lowest_votes = data[data['Votes'] == data['Votes'].min()]
lowest_votes_name = lowest_votes.iloc[0]['Restaurant Name']
lowest_votes_count = lowest_votes.iloc[0]['Votes']

print(f"Restaurant with the highest number of votes: {highest_votes_name} ({highest_votes_count} votes)")
print(f"Restaurant with the lowest number of votes: {lowest_votes_name} ({lowest_votes_count} votes)")
```

# 2. Analyze if there is a correlation between the number of votes and the rating of a restaurant

```python

# Select the relevant columns
data = data[['Votes', 'Aggregate rating']]

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=data)
plt.title('Correlation between Votes and Aggregate Rating')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)

# Calculate and print the correlation coefficient
correlation_coefficient = data['Votes'].corr(data['Aggregate rating'])
print(f"Correlation Coefficient: {correlation_coefficient}")

# Show the plot
plt.show()
```
