# TASK - 1 Top Cuisines
        1. Determine the top three most common cuisines in the dataset.
        2. Calculate the percentage of restaurants that serve each of the top cuisines.


```python
# Import library files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv('Dataset.csv')
```

```python
# Display the numbers of Rows and Column
data.shape
```

```python
#Dispaly the table
data
```

```python
# Count and display the number of Cuisines
No_of_cuisines = data['Cuisines'].value_counts()

No_of_cuisines
```

# 1. Determine the top three most common cuisines in the dataset.

```python
# Find and display the top 3 Cuisines
top3_cuisines = No_of_cuisines.head(3)
top3_cuisines
```

```python
# Create a bar plot to visualize the Top 3 Cuisines
plt.figure(figsize=(6, 3))
ax = top3_cuisines.plot(kind='bar')
plt.title('Top Three Cuisines in Restaurants')
plt.xlabel('Top 3 Cuisine')
plt.ylabel('No. of Cuisines')
plt.xticks(rotation=0)

# Display the values
for i, v in enumerate(top3_cuisines):
    ax.text(i, v + 1, str(v), ha='center')
plt.show()
```

# 2. Calculate the percentage of restaurants that serve each of the top cuisines.

```python
# Find and display the total number of cuisines
total_restaurants = len(data)
total_restaurants
```

```python
# Find and display the Percentage of cuisines
percentage = (top3_cuisines / total_restaurants) * 100
percentage
```

```python
plt.figure(figsize=(6, 3))
ax = percentage.plot(kind='barh')
plt.title('Percentage of Restaurants Serving Top Cuisines')
plt.xlabel('Percentage')
plt.ylabel('Cuisine')
for i, v in enumerate(percentage):
    ax.text(v + 1, i, f'{v:.2f}%', va='center')
plt.show()
```
