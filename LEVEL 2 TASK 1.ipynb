```python
# Import library files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('Dataset.csv')
data
```

```python
data.columns
```

# 1. Analyze the distribution of aggregate ratings and determine the most common rating range.

```python

# Create a histogram
n, bins, patches = plt.hist(data['Aggregate rating'], bins=10, edgecolor='k', color='green')

plt.xlabel('Rating Range')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate Ratings')

# Annotate the bars with their frequencies
for i in range(len(patches)):
    plt.text(patches[i].get_x() + patches[i].get_width() / 2, 
             patches[i].get_height() + 5,  # Adjust the offset for label placement
             str(int(n[i])), 
             ha='center', 
             va='bottom')

plt.show()
```

```python
# Find the bin with the highest frequency
most_common_range = bins[n.argmax()], bins[n.argmax() + 1]

print("Most common rating range:", most_common_range)

plt.show()

```

```python
mean_rating = data['Aggregate rating'].mean()
median_rating = data['Aggregate rating'].median()
std_deviation = data['Aggregate rating'].std()
min_rating = data['Aggregate rating'].min()
max_rating = data['Aggregate rating'].max()

print(f"Mean Rating: {mean_rating:.2f}")
print(f"Median Rating: {median_rating:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
print(f"Minimum Rating: {min_rating:.2f}")
print(f"Maximum Rating: {max_rating:.2f}")
```

# 2. Calculate the average number of votes received by restaurants.

```python
# Calculate the average number of votes
average_votes = data['Votes'].mean()

print(f"The average number of votes received by restaurants is: {average_votes:.2f}")
```
