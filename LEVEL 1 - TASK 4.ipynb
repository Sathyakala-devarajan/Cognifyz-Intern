# TASK - 4: Online Delivery
    1. Determine the percentage of restaurants that offer online delivery.
    2. Compare the average ratings of restaurants with and without online delivery.

```python
# Import library files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Dataset.csv')
data

```

```python
data.columns
```

# 1. Determine the percentage of restaurants that offer online delivery.

```python
online_delivery_counts = data['Has Online delivery'].value_counts()
online_delivery_counts
```

```python

# Calculate the total number of restaurants
total_restaurants = len(data)

# Calculate the percentage of restaurants that offer online delivery
percentage_with_online_delivery = (online_delivery_counts['Yes'] / total_restaurants) * 100

# Display the percentage
print(f"Percentage of restaurants that offer online delivery: {percentage_with_online_delivery:.2f}%")
```

```python
# Create a pie chart to visualize the percentage
plt.figure(figsize=(3, 3))
colors = ['skyblue', 'lightcoral']
labels = [f'Yes ({online_delivery_counts["Yes"]})', f'No ({online_delivery_counts["No"]})']
plt.pie([percentage_with_online_delivery, 100 - percentage_with_online_delivery], labels=labels, colors=colors, autopct='%1.2f%%', startangle=140)
plt.title('Percentage of Restaurants Offering Online Delivery')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```

# 2. Compare the average ratings of restaurants with and without online delivery.

```python

# Group the data by the 'Online Delivery' column and calculate the average rating for each group
average_ratings_by_delivery = data.groupby('Has Online delivery')['Aggregate rating'].mean()

# Display the average ratings for restaurants with and without online delivery
print("Average Ratings by Delivery Option:")
print(average_ratings_by_delivery)
```

```python
# Create a pie chart to visualize the average ratings
plt.figure(figsize=(3, 3))
colors = ['skyblue', 'lightcoral']
labels = [f'Yes ({average_ratings_by_delivery["Yes"]:.2f})', f'No ({average_ratings_by_delivery["No"]:.2f})']
plt.pie(average_ratings_by_delivery, labels=labels, colors=colors, autopct='%1.2f%%', startangle=140)
plt.title('Average Ratings for Restaurants Online Delivery')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

plt.show()
```
